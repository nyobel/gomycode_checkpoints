## Data Preprocessing

```python
# handle null values
df.isna().sum()
```

```python
df_clean =df.copy()
```

```python
# 1. Calculate the percentage of missing values per column
missing_percentage = df_clean.isna().mean() * 100
print("Percentage of missing values per column:")
print(missing_percentage.sort_values(ascending=False))
```

```python
# Separate numerical and categorical columns
num_cols = df_clean.select_dtypes(include=['float64', 'int64']).columns.tolist()
cat_cols = df_clean.select_dtypes(include=['object']).columns.tolist()

# For numerical columns: fill missing values with the median
for col in num_cols:
    median_val = df_clean[col].median()
    df_clean[col].fillna(median_val, inplace=True)
    print(f"Filled missing values in numerical column '{col}' with median: {median_val}")

# For categorical columns: fill missing values with the mode (most common value)
for col in cat_cols:
    if not df_clean[col].mode().empty:
        mode_val = df_clean[col].mode()[0]
    else:
        mode_val = "Missing"  # fallback if mode cannot be calculated
    df_clean[col].fillna(mode_val, inplace=True)
    print(f"Filled missing values in categorical column '{col}' with mode: {mode_val}")

# 4. Check remaining missing values to validate our approach
print("\nMissing values after cleaning:")
print(df_clean.isna().sum())
```

```python
# handle duplicates
df_clean.duplicated().sum()
```

```python
df_clean.drop(["user_id", "MRG"], axis = 1, inplace=True)
df_clean.columns
```

## Encoding

```python
from sklearn.preprocessing import LabelEncoder

label_encoders = {}
for column in df_clean.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    df_clean[column] = le.fit_transform(df_clean[column])
    label_encoders[column] = le
```

## Model Training

```python
# select features and target variable
X = df_clean.drop('CHURN', axis=1)
y = df_clean['CHURN']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify = y, random_state=42)
## "stratify = y", to make sure the districution of the target "CHURN" are evenly distributed in the train and test sets
```

```python
# Train a Random Forest model
model = RandomForestClassifier(n_estimators=100, random_state=42)

model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report

# Predict on the test set
y_pred = model.predict(X_test)

# Evaluate performance
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

accuracy, report
```

```python
import joblib

# Save best model
joblib.dump(model, 'churn_model.pkl')

# Save label encoders (critical for categorical features)
joblib.dump(label_encoders, 'label_encoders.pkl')

# Save feature names (important for column order)
joblib.dump(X_train.columns.tolist(), 'feature_names.pkl')
```
