```python
#handling missing values
# Impute missing values in counter_number with its median
df['counter_number'].fillna(df['counter_number'].median(), inplace=True)

# Impute missing values in reading_remarque with its median
df['reading_remarque'].fillna(df['reading_remarque'].median(), inplace=True)
```

```python
df.isnull().sum()
```

#why imputation to handle missing values.
- counter_number column seems important (likely an identifier for the meter).

- Data Retention: Imputation keeps all 4.48M rows intact, ensuring no loss of potentially valuable data.
- Low Missing Percentage: With only 0.001% missing in counter_number and 0.1% in reading_remarque, the imputed values are unlikely to skew analysis of the data significantly.
